```python
# Import statements for this file
import bs4
import requests
import pandas as pd 
import numpy as np
from bs4 import BeautifulSoup
```

```python
# Web scrapping initialization for PA zip code webpage 
root = "https://www.zipcodestogo.com/Pennsylvania/"
html = requests.get(root)
soup = BeautifulSoup(html.content,'html.parser')
elements = soup.select('tr a')
```

```python
# Strip extra spaces from 'tr a' that we grabbed above 
list=[e.text.strip() for e in elements]
num=0
zip=[]
county=[]
# Loop through elements in list and append them to the correct list
for i in list:
    if (num%3==1 and num<=6571):
        zip.append(i)
    if (num%3==2 and num<=6572):
        county.append(i)
    num=num+1
```

```python
# Create data frame of counties and zip codes scraped from website
headers = ['ZipCode', 'County']
df = pd.DataFrame(columns=headers)
point=0
num=1
repeat=[]
# Append data to df and repeat list
for i in zip:
    if (county[point] in repeat)==False:
        df.loc[num]=[i, county[point]]
        repeat.append(county[point])
        num+=1
    point+=1
```

```python
# Create data frame of cities and zip codes scraped from website
headers = ['ZipCode', 'City']
df1 = pd.DataFrame(columns=headers)
num=0
repeat=[]
point=1
# Append data to df and repeat list
for i in elements:
    if(num%3==1 and num<=6547):
        output=i['href'].split('/')
        if (output[3] in repeat)==False:
            df1.loc[point]=[output[5], output[3]]
            repeat.append(output[3])
            point+=1
    num=num+1
```

```python
# Save data to files
df.to_csv('County_Zip.csv', sep=',', encoding='utf-8')
df1.to_csv('City_Zip.csv', sep=',', encoding='utf-8')
```
